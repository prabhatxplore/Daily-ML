```python
import pandas as pd
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("/kaggle/input/covidtoycsv/covid_toy.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size =0.2)
```

```python
# simple imputer on fever column train data
si = SimpleImputer()
X_train_fever  = si.fit_transform(X_train[['fever']])

# simple imputer on fever col test data
X_test_fever = si.transform(X_test[['fever']])

print(type(X_train_fever))
```

```python
# Ordinal Encoding -> cough
oe = OrdinalEncoder(categories= [['Mild','Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']])

# test data
X_test_cough = oe.fit_transform(X_test[['cough']])
X_train_cough.shape
print(type(X_train_cough))
```

```python
# One Hot Encoding ->
oh = OneHotEncoder(drop='first',sparse_output=False)
X_train_genci = oh.fit_transform(X_train[['gender','city']])


# test data
X_test_genci = oh.transform(X_test[['gender','city']])

X_train_genci.shape
print(type(X_train_genci.shape))
```

```python
# Extracting Age 
X_train_age  = X_train.drop(columns=['gender','fever','city','cough'])


# test data
X_test_age = X_test[['age']]
print(type(X_train_age.shape))
```

```python
X_train_transformed = np.concatenate((X_train_genci,X_train_age,X_train_fever,X_train_cough),axis=1)
X_test_transformed = np.concatenate((X_test_genci,X_test_age,X_test_fever,X_test_cough),axis=1)
```

# Column Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder = 'passthrough')
```
