```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state =4)
```

```python
from sklearn.linear_model import SGDRegressor,Ridge
```

```python
sgd = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,
                  learning_rate='constant',alpha=0.001)
sgd.fit(X_train,y_train)
y_pred = sgd.predict(X_test)
print(r2_score(y_test,y_pred))
print(sgd.intercept_)
print(sgd.coef_)
```

```python
reg = Ridge(alpha=0.001, max_iter=500,solver='sparse_cg')
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_) 
```

```python
class GDRidge:
    def __init__(self,alpha=0.01,learning_rate=0.1,epochs=100):
        self.alpha = alpha
        self.lr = learning_rate 
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X_train,y_train):
        self.coef_ = np.ones(X_train.shape[1])
        self.intercept_ = 0
        y_train = y_train.reshape(-1,1)
        theta = np.insert(self.coef_,0,self.intercept_).reshape(-1,1)
        X_train = np.insert(X_train,0,1,axis=1)
        print(theta.shape)
        print(X_train.shape)
        print(y_train.shape)
        for i in range(self.epochs):
            theta_d = X_train.T.dot(X_train.dot(theta)) - np.dot(X_train.T,y_train) + self.alpha * theta
    
            theta = theta - self.lr * theta_d

        self.intercept_ = theta[0]
        self.coef_ = theta[1:]
        print(self.intercept_)
        print(self.coef_)
    def predict(self,X_test):
        return X_test.dot(self.coef_)+self.intercept_
```

```python
gdr  = GDRidge(alpha=0.001,learning_rate=0.1,epochs=100)
gdr.fit(X_train,y_train)
r2_score(y_test,gdr.predict(X_test))
```
