```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/placementcsv/placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df.cgpa,df.package)
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.show()
```

```python
X   = df.iloc[:,:1]
y = df.iloc[:,1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
plt.scatter(X_train,y_train,label = 'training data',alpha=0.5)
plt.plot(X_train,lr.predict(X_train),label = 'predicted',color ='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

```

```python
y_pred = lr.predict(X_test)
```

```python
print('MAE ',mean_absolute_error(y_test,y_pred))
```

```python
# MAE 
np.mean(np.abs(y_test-y_pred))
```

```python
print('MSE',mean_squared_error(y_test,y_pred))
```

```python
# MSE 
print(np.mean((y_test-y_pred)**2))
print(np.sqrt(np.mean((y_test-y_pred)**2)))
```

```python
print('RMSE',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# R2 score
1-np.mean((y_test-y_pred)**2)/np.mean((y_test-np.mean(y_test))**2)
```

```python
print('R2_score',r2_score(y_test,y_pred))
r2 = r2_score(y_test,y_pred)
```

```python
print('Adjusted R2 :  ',1-(((1-r2) * (X_test.shape[0]-1))/(X_test.shape[0]-1-X_test.shape[1])))
```
