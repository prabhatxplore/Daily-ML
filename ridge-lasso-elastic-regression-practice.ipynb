```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score , root_mean_squared_error
```

```python
df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv',index_col=0)
sub_df =pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv',index_col=0)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size = 0.2,random_state=42)
```

```python
num  = X_train.select_dtypes(include='number')
X_test =X_test.select_dtypes(include= 'number')
sub_df_num = sub_df.select_dtypes(include='number')
X_test.shape
```

```python
# extract the cols names to impute
impute_cols = num.columns[num.isna().any()].tolist()
impute_cols_sub = sub_df_num.columns[sub_df_num.isna().any()].tolist()
impute_cols_scal = list(set(impute_cols+impute_cols_sub))
print(impute_cols_scal)
```

```python

# drop the categorical cols
num_drop_impute_cols = num.drop(columns=impute_cols_sub)
num_droptest_impute_cols = X_test.drop(columns = impute_cols_sub)
sub_drop_impute_cols = sub_df_num.drop(columns=impute_cols_sub)
print(num_droptest_impute_cols.shape)
print(num_drop_impute_cols.shape)
print(sub_drop_impute_cols.shape)
```

```python
# impute the missing value in train and in test dataset
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values = np.nan,strategy='mean')
print(impute_cols_scal)
X_train_impute = impute.fit_transform(num[impute_cols_scal])
X_test_impute = impute.transform(X_test[impute_cols_scal])
sub_df_impute = impute.transform(sub_df[impute_cols_scal])

print(X_train_impute.shape)
print(X_test_impute.shape)
print(sub_df_impute.shape)
```

```python
# combine the imputed cols with the dropped cols from above
X_train_df = np.hstack((num_drop_impute_cols.values,X_train_impute)) 
X_test_df = np.hstack((num_droptest_impute_cols.values,X_test_impute))
sub_df_num = np.hstack((sub_drop_impute_cols.values,sub_df_impute))
```

```python
print(X_train_df.shape)
print(X_test_df.shape)
print(sub_df_num.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_df = scaler.fit_transform(X_train_df)
X_test_df = scaler.transform(X_test_df)
sub_df_num = scaler.transform(sub_df_num)
```

```python
print(pd.DataFrame(X_train_df).head())
print(pd.DataFrame(X_test_df).head())
print(pd.DataFrame(sub_df_num).head())
```

# Linear Regression

```python
reg = LinearRegression()
reg.fit(X_train_df,y_train)
y_pred = reg.predict(X_test_df)
print(reg.intercept_)
print(reg.coef_)
print('R2 Score : ',r2_score(y_test,y_pred))
print(' RootMeanSquared Log : ',root_mean_squared_error(np.log(y_test),np.log(y_pred)))
```

# Ridge Regression

```python
ridg = Ridge(alpha=13,solver='sparse_cg',max_iter=1000)
ridg.fit(X_train_df,y_train)
print(ridg.intercept_)
print(ridg.coef_)

y_pred = ridg.predict(X_test_df)
print('R2 Score : ',r2_score(y_test,y_pred))
print(' RootMeanSquared Log : ',root_mean_squared_error(np.log(y_test),np.log(y_pred)))
```

# Lasso Regression

```python
lass  = Lasso(alpha=25,max_iter=1000)
lass.fit(X_train_df,y_train)
print(lass.intercept_)
print(lass.coef_)
y_pred = lass.predict(X_test_df)
print('R2 Score : ',r2_score(y_test,y_pred))
print(' RootMeanSquared Log : ',root_mean_squared_error(np.log(y_test),np.log(y_pred))) 
```

# Elastic Regression

```python
elastic = ElasticNet(alpha =0.01,l1_ratio=0.7,max_iter=10000)
elastic.fit(X_train_df,y_train)
print(elastic.intercept_)
print(elastic.coef_)
y_pred = elastic.predict(X_test_df)
print('R2 Score : ',r2_score(y_test,y_pred))
print(' RootMeanSquared Log : ',root_mean_squared_error(np.log(y_test),np.log(y_pred)))
```
