```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/iris/Iris.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df['Species'] = LabelEncoder().fit_transform(df['Species'].values)
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size = 0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
cl1 = LogisticRegression()
cl2 = DecisionTreeClassifier()
```

```python
cl1.fit(X_train,y_train)
cl2.fit(X_train,y_train)
```

```python
y_pred1 = cl1.predict(X_test)
y_pred2 = cl2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,precision_score,recall_score
```

```python
print('Accuracy',accuracy_score(y_test,y_pred1))
print('Accuracy',accuracy_score(y_test,y_pred2))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred1,average = 'weighted')
```

```python
print('F1 score ',f1_score(y_test,y_pred2,average = 'macro'))
print('Recall Score ',recall_score(y_test,y_pred2,average=None))
print('Precision score ',precision_score(y_test,y_pred2,average=None))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred1))
print('\n \n')
print(classification_report(y_test,y_pred2))
```

# Precision says out of all true prediction how many of them are actually true

# Recall says out of all actual true how many of them are correctly classified

# F1 - score is the harmonic mean of the recall and precision. it tells whoever is the less i give that value
